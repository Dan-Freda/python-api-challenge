```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import json
import time
import datetime
from scipy.stats import linregress
from pprint import pprint

# Import API key
from config import api_key

# Incorporated citipy to determine city based on latitude and longitude
from citipy import citipy

# Output File (CSV)
output_data_file = "output_data/cities.csv"

# Range of latitudes and longitudes
lat_range = (-90, 90)
lng_range = (-180, 180)
```

## Generate Cities List

```python
# List for holding lat_lngs and cities
lat_lngs = []
cities = []

# Create a set of random lat and lng combinations
lats = np.random.uniform(lat_range[0], lat_range[1], size=1500)
lngs = np.random.uniform(lng_range[0], lng_range[1], size=1500)
lat_lngs = zip(lats, lngs)

# Identify nearest city for each lat, lng combination
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    # If the city is unique, then add it to a our cities list
    if city not in cities:
        cities.append(city)

# Print the city count to confirm sufficient count
len(cities)
```

### Perform API Calls
* Perform a weather check on each city using a series of successive API calls.
* Include a print log of each city as it'sbeing processed (with the city number and city name).


```python
city_name_list = []
country_list = []
lat_list = []
lng_list = []
max_temp_list = []
humidity_list = []
cloudiness_list = []
wind_speed_list = []
date_list = []
index_counter = 0
set_counter = 1
```

```python
print("Beginning Data Retrieval ")
print("-----------------------------")

base_url = "http://api.openweathermap.org/data/2.5/weather?"
units = "imperial"
query_url = f"{base_url}appid={api_key}&units={units}&q="


# For each city name in cities list, do below things...
for index, city in enumerate(cities, start = 1):
    try:
        response = requests.get(query_url + city).json()
        city_name_list.append(response["name"])
        cloudiness_list.append(response["clouds"]["all"])
        country_list.append(response["sys"]["country"])
        date_list.append(response["dt"])
        humidity_list.append(response["main"]["humidity"])
        lat_list.append(response["coord"]["lat"])
        lng_list.append(response["coord"]["lon"])
        max_temp_list.append(response['main']['temp_max'])
        wind_speed_list.append(response["wind"]["speed"])
        if index_counter > 49:
            index_counter = 0
            set_counter = set_counter + 1
    
        else:
            index_counter = index_counter + 1
            
        print(f"Processing Record {index_counter} of Set {set_counter} : {city}") 
  
    except(KeyError, IndexError):
        print("City not found. Skipping...")

print("-----------------------------")
print("Data Retrieval Complete")
print("-----------------------------")
```

### Convert Raw Data to DataFrame
* Export the city data into a .csv.
* Display the DataFrame

```python
# Create a Dataframe comprised of the results in response to the API calls
weather_df = pd.DataFrame({
                "City" : city_name_list,
                "Country" : country_list,
                "Lat" : lat_list,
                "Lng" : lng_list,
                "Max Temp (F)" : max_temp_list,
                "Humidity" : humidity_list,
                "Cloudiness" : cloudiness_list,
                "Wind Speed (mph)" : wind_speed_list,
                "Date" : date_list,
})
```

```python
# Perform a count to verify how many cities return in response to the API call
weather_df.count()
```

```python
# Display the DataFrame
weather_df
```

```python
# Export the city data into a .csv
weather_df.to_csv("output_data/cities.csv", index = False)
```

## Linear Regression

```python
# Create Northern and Southern Hemisphere datasets for this section of comparative analysis
northern_hem = weather_df.loc[weather_df["Lat"] >= 0]
northern_hem.reset_index(inplace=True)

southern_hem = weather_df.loc[weather_df["Lat"] < 0]
southern_hem.reset_index(inplace=True)
```

```python
# Define the function that will be used to run the linear regression on each relationship, plotting, and storing image file

def plotLinearRegression(xval, yval, xlbl, ylbl, lblpos, ifig):
    (slope, intercept, rvalue, pvalue, stderr) = linregress(xval, yval)
    print(f"The r-squared is: {rvalue}")
    regress_values = xval * slope + intercept
    line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))

    plt.scatter(xval, yval)
    plt.plot(xval, regress_values, "r-")
    plt.annotate(line_eq,lblpos,fontsize=15,color="red")
    plt.xlabel(xlbl)
    plt.ylabel(ylbl)
    plt.savefig(f"output_data/fig{ifig}.png")
    plt.show()
```

##  Northern Hemisphere - Max Temp vs. Latitude Linear Regression

```python
# Create relevant titles
xlbl = "Lat"
ylbl = "Max Temp (F)"

# Label position altered for better clarity in visualization
lblpos = (0,25)
plotLinearRegression(northern_hem[xlbl], northern_hem[ylbl], xlbl, ylbl, lblpos, 1)
```

#### Analysis:

There is a strong negative correlation betwwen max temperature and latitude. This is evidenced by the high r-squared value and the negative slope of the line. As temperatures increase in value, latitude values - distance from the equator - decrease. The inverse is true as well. 

##  Southern Hemisphere - Max Temp vs. Latitude Linear Regression

```python
# Create relevant titles
xlbl = "Lat"
ylbl = "Max Temp (F)"

# Label position altered for better clarity in visualization
lblpos = (-55, 80)
plotLinearRegression(southern_hem[xlbl], southern_hem[ylbl], xlbl, ylbl, lblpos, 2)
```

#### Analysis:

Similar to the northern hemisphere, there is a strong linear relationship and correlation between max temperature and latitude. However, there is positive relationship in this case: as max temperatures increase in value, so do latitude values. Southern hemisphere latitude values are categorized in negative, so the relationship is essentially the same: temperatures increase as distance to the equator decrease.
